# Carga y descrga de un condensador

La carga de un condensador esta dada por:

$q=\epsilon C(1-e^{-t/RC})$ 

El flujo de carga por unidad de tiempo (o sea la corrinte) sera:

$i\dfrac{dq}{dt}=i=\dfrac{\epsilon}{R}e^{-t/RC}$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#funcion carga
#e=volatje
#C=capacitancia
#R=reisitencia
#t=tiempo
def cargaQ(e,C,R,t):
    return e*C*(1-np.exp(-t/(R*C)))
```

```python
t = np.linspace(0, 2000, 50000)
Q=cargaQ(5,0.06,10000,t)
plt.plot(t,Q)
plt.ylabel('Carga')
plt.xlabel('t (s)')
plt.show()
```

```python
#funcion corriente
def cargaI(e,C,R,t):
    return (e/R)*np.exp(-t/(R*C))

I=cargaI(5,0.06,10000,t)
plt.plot(t,I,'r')
plt.ylabel('I(amp)')
plt.xlabel('t (s)')
plt.show()
```

La descarga de un condensador esta dada por:

$q=\epsilon Ce^{-t/RC})$ 

El flujo de carga por unidad de tiempo (o sea la corriente) sera:

$i\dfrac{dq}{dt}=i=-\dfrac{\epsilon}{R}e^{-t/RC}$

```python
#funcion descarga
#e=volatje
#C=capacitancia
#R=reisitencia
#t=tiempo
def descargaQ(e,C,R,t):
    return e*C*(np.exp(-t/(R*C)))

desQ=descargaQ(5,0.06,10000,t)
plt.plot(t,desQ)
plt.ylabel('descarga')
plt.xlabel('t (s)')
plt.show()
```

```python
#funcion corriente descarga
def descargaI(e,C,R,t):
    return -(e/R)*np.exp(-t/(R*C))

desI=descargaI(5,0.06,10000,t)
plt.plot(t,desI)
plt.ylabel('I(amp)')
plt.xlabel('t (s)')
plt.show()
```
